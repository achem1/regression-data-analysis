# Regression Plots

Seaborn has many built-in capabilities for regression plots, we discuss regression until the machine learning section of the course, so we will only cover the **lmplot()** function for now.

**lmplot** allows you to display linear models, but it also conveniently allows to split up those plots based of features, as well as coloring the hue based of features.

Let see how regression works:

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

### Working with Markers

lmplot kwargs get through  **regplot** which is more general form of lmplot(). regplot has a scatter_kws parameter that get to plt.scatter. you want to set the s parameter in the dictionary, which corresponds to the squared markersize. In other word you end up passing a dictionary with the matplotlib arguments, in this case, s for size of a scatter plot.

```python
# http://matplotlib.org/api/markers_api.html
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid

you can add variable separation through columns and rows with the use of a grid. 

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm')
```

## Aspect and Size

Seaborn figures can have their size and aspect ratio adjusted parameter:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
           aspect = 0.6)
```
